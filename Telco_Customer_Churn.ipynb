```python
import plotly.graph_objects as go

# Veri setini yükleyelim
df = pd.read_csv('Telco-Customer-Churn.csv')

# Sayısal sütunları kontrol edelim ve gerekli dönüşümleri yapalım
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df.dropna(inplace=True)
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
```

```python
# Month-to-Month sözleşmesi olan müşterileri filtreleme
df_month_to_month = df[df['Contract'] == 'Month-to-month']

# Korelasyon matrisini hesaplayalım (Month-to-Month müşteriler için)
correlation_matrix_month_to_month = df_month_to_month[['tenure', 'MonthlyCharges', 'TotalCharges']].corr()
```

```python
# One Year sözleşmesi olan müşterileri filtreleme
df_one_year = df[df['Contract'] == 'One year']

# Korelasyon matrisini hesaplayalım (One Year müşteriler için)
correlation_matrix_one_year = df_one_year[['tenure', 'MonthlyCharges', 'TotalCharges']].corr()
```

```python
# Two Year sözleşmesi olan müşterileri filtreleme
df_two_year = df[df['Contract'] == 'Two year']

# Korelasyon matrisini hesaplayalım (Two Year müşteriler için)
correlation_matrix_two_year = df_two_year[['tenure', 'MonthlyCharges', 'TotalCharges']].corr()
```

```python
fig_month_to_month = go.Figure(data=go.Heatmap(
                   z=correlation_matrix_month_to_month.values,
                   x=correlation_matrix_month_to_month.columns,
                   y=correlation_matrix_month_to_month.columns,
                   colorscale='Viridis'))

fig_month_to_month.update_layout(title='Correlation Matrix (Month-to-Month)', autosize=False, width=500, height=500)
fig_month_to_month.show()
```

```python
fig_one_year = go.Figure(data=go.Heatmap(
                   z=correlation_matrix_one_year.values,
                   x=correlation_matrix_one_year.columns,
                   y=correlation_matrix_one_year.columns,
                   colorscale='Viridis'))

fig_one_year.update_layout(title='Correlation Matrix (One Year)', autosize=False, width=500, height=500)
fig_one_year.show()
```

```python
fig_two_year = go.Figure(data=go.Heatmap(
                   z=correlation_matrix_two_year.values,
                   x=correlation_matrix_two_year.columns,
                   y=correlation_matrix_two_year.columns,
                   colorscale='Viridis'))

fig_two_year.update_layout(title='Correlation Matrix (Two Year)', autosize=False, width=500, height=500)
fig_two_year.show()
```

```python
# Sayısal sütunları seç
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
correlation_matrix = df[numerical_columns].corr()

# Korelasyon matrisini göster
print(correlation_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Aylık ücretleri kategorik aralıklara grupla
df['MonthlyChargesCategory'] = pd.cut(df['MonthlyCharges'], bins=5)

# Ortalama churn oranlarını hesapla
churn_rates = df.groupby('MonthlyChargesCategory')['Churn'].mean().reset_index()

# Çubuk grafiğini oluştur
plt.figure(figsize=(10, 6))
sns.barplot(x='MonthlyChargesCategory', y='Churn', data=churn_rates)
plt.title('Average Churn Rate by Monthly Charges')
plt.xlabel('Monthly Charges Category')
plt.ylabel('Average Churn Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
# Aylık ücretleri kategorik aralıklara grupla
df['MonthlyChargesCategory'] = pd.cut(df['MonthlyCharges'], bins=5)

# Ortalama hizmet süresini hesapla
tenure_means = df.groupby('MonthlyChargesCategory')['tenure'].mean().reset_index()

# Çubuk grafiğini oluştur
plt.figure(figsize=(10, 6))
sns.barplot(x='MonthlyChargesCategory', y='tenure', data=tenure_means)
plt.title('Average Tenure by Monthly Charges')
plt.xlabel('Monthly Charges Category')
plt.ylabel('Average Tenure')
plt.xticks(rotation=45)
plt.show()
```

```python
# Aylık ücretlere göre kutu grafiği oluşturalım
plt.figure(figsize=(10, 8))
sns.boxplot(x='PaymentMethod', y='MonthlyCharges', data=df)
plt.title('Monthly Charges by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Monthly Charges')
plt.show()
```
